```python
#Importing Mmodules and Dependencies
%%capture
!pip install PyMuPDF
import spacy
import sys,fitz
import re
```

```python
#Getting the pdf and writing text to a txt file
def pdfreader(pdfpath,textpath):
  doc=fitz.open(pdfpath)
  text=""
  for page in doc:
    text=text+str(page.getText())
  file1 = open(textpath,"w")
  file1.write(text)
  file1.close()
  return textpath
```

```python
#Extracting Name, Email and Phone Number from the .txt file
def basicinfo(pdfpath,textpath):
  txtp=pdfreader(pdfpath,textpath)
  with open(txtp, 'r') as file:
    tx = file.read().replace('\n', ' ')
  names=tx.split()
  name=names[0]+" "+names[1]
  mails= re.search(r'[\w.+-]+@[\w-]+\.[\w.-]+', tx)
  mail=mails.group(0)
  phones=re.findall(r'([0-9][0-9 .\\(\)]{8,}[0-9])', tx)
  phone=phones[0]
  return tx,name,mail,phone
```

```python
#Extracting other important info using pretrained model developed using spacy.
def extractor(pdfpath,textpath):
  nlpmodel=spacy.load('/content/drive/MyDrive/KA-Innovations/NLP/nlp_model')# Loading the saved model.
  txt,n,m,p=basicinfo(pdfpath,textpath)
  doc=nlpmodel(txt)
  extracted_data={"CV":"CV"}
  for ent in doc.ents:
  #print(f'{ent.label_.upper():{30}}-{ent.text}')
    extracted_data[ent.label_.upper()]=ent.text
  return n,m,p,extracted_data
  
```

```python
#Function which executes operations and logic for answer the qustions entered by the user.
def infoandquestion(pdfpath,textpath):
  n,m,p,extracted_data=extractor(pdfpath,textpath)
  print(n)
  print(m)
  print(p)

  
  val = str(input("Enter your question: "))
  

  if "name"  in val.lower():
    print(extracted_data["NAME"])
    print("Try another question.")
  elif "college" in val.lower() or "university" in val.lower():
    print(extracted_data["COLLEGE NAME"])
    print("Try another question.")
  elif  "skills" in val.lower():
    print(extracted_data["SKILLS"])
    print("Try another question.")
  elif  "degree" in val.lower():
    print(extracted_data["DEGREE"])
    print("Try another question.")
  elif "position" in val.lower() or "designation" in val.lower():
    print(extracted_data["DESIGNATION"])
    print("Try another question.")
  elif "location" in val.lower() or "home" in val.lower():
    print(extracted_data["LOCATION"])
    print("Try another question.")
  else:
    print("Unrelated Question, Try Again.")
```
